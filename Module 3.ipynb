## Project Title:
Interpolating Lung Fibrosis Severity from Limited Biopsy Imaging Data



## Project Goal:
This project aims to develop an image analysis and interpolation pipeline that predicts the extent of lung fibrosis at varying depths within the lung using a limited set of biopsy images. Because clinical biopsies cannot be taken continuously throughout the lung, this project uses interpolation methods to estimate fibrosis severity at unmeasured depths. The ultimate goal is to support the development of safer and more informative lung biopsy devices by providing clinicians with a continuous estimate of fibrosis burden based on sparse imaging data.

## Disease Background: 
Disease Background: Idiopathic Pulmonary Fibrosis (IPF)

1. Prevalence and Incidence

Idiopathic Pulmonary Fibrosis (IPF) is a rare, chronic interstitial lung disease that primarily affects older adults, with most diagnoses occurring after age 60. In the United States, prevalence is estimated at approximately 13 to 20 cases per 100,000 individuals, and incidence increases sharply with age. Although IPF is classified as a rare disease, its prevalence has increased due to improved diagnostic imaging techniques and an aging population.

2. Economic Burden

IPF imposes a significant economic burden on both patients and healthcare systems. Costs arise from long term medication use, repeated hospitalizations, supplemental oxygen therapy, pulmonary rehabilitation, and frequent clinical monitoring. Indirect costs, including loss of income, reduced productivity, and disability, further increase the overall financial impact of the disease.

3. Risk Factors (Genetic and Lifestyle)

The development of IPF is influenced by a combination of genetic predisposition and environmental or lifestyle factors. Genetic mutations affecting telomere length maintenance and surfactant protein production increase susceptibility. Lifestyle and environmental risk factors include cigarette smoking, occupational exposure to dust or metal particles, air pollution, and potentially vaping related lung injury, which may contribute to earlier disease onset.

4. Societal Determinants

Social determinants of health significantly influence IPF outcomes. Access to specialized pulmonary care, socioeconomic status, geographic location, and health insurance coverage affect the timeliness of diagnosis and treatment. Patients in rural or underserved communities may experience delayed diagnosis or limited access to antifibrotic medications and transplant centers, resulting in worse prognoses.

5. Symptoms

Symptoms of IPF typically progress gradually and are often nonspecific in early stages. Common symptoms include persistent dry cough, increasing shortness of breath during exertion, fatigue, and reduced exercise tolerance. As the disease advances, lung stiffness worsens, leading to severe breathing difficulty and the eventual need for continuous supplemental oxygen.

6. Diagnosis

Diagnosis of IPF relies on a combination of clinical evaluation, pulmonary function testing, and imaging. High resolution computed tomography scans are essential for identifying hallmark features such as honeycombing patterns within the lung. Pulmonary function tests often reveal reduced forced vital capacity, reflecting decreased lung compliance. In some cases, lung biopsy is required to confirm the diagnosis.

7. Standard of Care Treatments

There is currently no cure for IPF, and standard treatments focus on slowing disease progression and managing symptoms. Antifibrotic medications such as pirfenidone and nintedanib are commonly prescribed to reduce the rate of lung function decline. Supportive therapies include oxygen supplementation and pulmonary rehabilitation. Lung transplantation remains the only definitive treatment but is limited by patient eligibility and donor availability.

8. Disease Progression and Prognosis

IPF is a progressive and irreversible disease characterized by continuous accumulation of scar tissue within the lungs. Excess collagen deposition leads to reduced lung elasticity, loss of alveoli, and impaired gas exchange. Prognosis is generally poor, with a median survival of approximately three to five years following diagnosis, though progression rates vary between individuals.

9. Continuum of Care Providers

Management of IPF requires a multidisciplinary care team. Pulmonologists coordinate diagnosis and treatment, radiologists interpret imaging studies, and pathologists analyze tissue samples when biopsies are performed. Respiratory therapists assist with oxygen therapy and rehabilitation, while transplant specialists evaluate and manage patients eligible for lung transplantation. Collaborative care is essential for optimal disease management.

10. Biological Mechanisms

Biologically, IPF results from abnormal wound healing responses within the lung interstitial tissue, the space between alveolar epithelial cells and blood capillaries. Repeated epithelial injury leads to excessive activation of myofibroblasts, which produce large amounts of collagen and fail to undergo normal apoptosis. This excessive collagen accumulation causes permanent scarring, loss of alveolar structure, reduced lung compliance, and impaired oxygen diffusion.

11. Clinical Trials and Next Generation Therapies

Ongoing research into IPF includes clinical trials exploring new antifibrotic agents, anti inflammatory therapies, stem cell treatments, and precision medicine approaches. Advances in imaging, biopsy device technology, and computational modeling, such as interpolation methods to estimate fibrosis severity, play an important role in improving early detection, disease monitoring, and personalized treatment strategies.

## Data-Set Description: 

"Unpublished data was collected by the Peirce-Cottler Lab (Dept. of Biomedical Engineering) and Kim Lab (Division of Pulmonary and Critical Care) at the University of Virginia School of Medicine. 

Data Source

The dataset used in this project consists of unpublished data collected by the Peirce Cottler Lab in the Department of Biomedical Engineering and the Kim Lab in the Division of Pulmonary and Critical Care at the University of Virginia School of Medicine. The data were generated as part of a preclinical research study investigating lung fibrosis using a mouse model. This dataset was provided specifically for educational and research purposes within this course.

Experimental Model

Lung fibrosis was induced in mice using a bleomycin injury model, which is a well-established experimental method for studying idiopathic pulmonary fibrosis. Bleomycin is an antibiotic and chemotherapeutic agent that causes lung tissue damage as a known side effect. Researchers leverage this effect to model fibrotic lung disease in mice. After bleomycin administration, mice were allowed to recover for three weeks to allow fibrosis to develop. The animals were then humanely sacrificed, and lung tissue was harvested for analysis.

Tissue Preparation and Imaging

After harvesting, mouse lungs were fixed using paraformaldehyde and embedded in either gel or wax to preserve tissue structure. The lungs were then sliced into thin sections using a cryotome or microtome. Importantly, slices were taken in the transverse plane, allowing images to be collected at different depths from the top to the bottom of the lung. Each tissue slice was mounted onto a glass microscope slide for further processing.

Immunostaining and Signal Selection

Lung tissue sections were immunostained using fluorescently labeled antibodies targeting three proteins of interest: desmin, smooth muscle alpha actin, and CD31. Desmin labels myofibroblasts, the primary cells responsible for producing fibrotic scar tissue in the lung. Smooth muscle alpha actin labels smooth muscle cells in larger blood vessels, and CD31 labels endothelial cells present in all blood vessels, including capillaries. For this project, only the desmin signal was used, as it directly represents fibrotic activity.

Image Processing and Data Format

Fluorescent microscopy was used to capture high resolution images of the immunostained lung sections. The desmin signal was extracted and converted into black and white images to simplify analysis. In these processed images, white pixels represent fibrotic lesions where myofibroblasts are present, while black pixels represent healthy lung tissue. A total of 78 black and white images were collected, each corresponding to a specific depth within the lung.

Measurement Units and Depth Information

Each image in the dataset corresponds to a known physical depth within the lung, measured in micrometers. Images span from near the top of the lung down to approximately 10,000 micrometers in depth. The primary quantitative measurement derived from the images is the proportion of fibrotic tissue, calculated based on the number or fraction of white pixels in each image. These fibrosis measurements as a function of lung depth serve as the input for interpolation, allowing estimation of fibrosis at depths where images were not directly collected.

## Conclusions and Ethical Implications: 

This project demonstrates that image analysis combined with interpolation techniques can be used to estimate the extent of lung fibrosis at depths where biopsy samples are not directly available. By quantifying fibrotic tissue from immunostained lung images and modeling fibrosis as a function of depth, the analysis addresses a key clinical challenge: the inability to safely sample lung tissue continuously throughout the organ. The results suggest that interpolation can provide a reasonable approximation of fibrosis distribution, which may support improved diagnostic insight and inform the design of safer lung biopsy devices.

There are important ethical implications associated with these findings. Estimating disease severity using computational methods introduces the risk of misinterpretation if predictions are treated with the same confidence as direct measurements. In a clinical setting, inaccurate interpolation could influence diagnostic or treatment decisions, potentially leading to overestimation or underestimation of disease severity. Ethical use of such methods therefore requires transparency about uncertainty, clear communication of limitations, and validation against ground truth data. Additionally, the dataset is derived from animal models, which raises ethical considerations regarding animal research. While animal models are essential for studying disease mechanisms and developing medical technologies, their use must be justified by potential benefits to human health and conducted under strict ethical and regulatory oversight.


## Limitations and Future Work: 
This project has several limitations that should be considered when interpreting the results. First, the dataset consists of a limited number of images collected from a single fibrotic mouse lung model. As a result, the interpolation may not capture the full biological variability present across different subjects, disease severities, or lung regions. Second, the analysis relies on two dimensional black and white images of a single biomarker, desmin, which represents myofibroblast activity but does not capture other important aspects of lung structure or fibrosis progression. Additionally, interpolation assumes relatively smooth changes in fibrosis with depth, which may not fully reflect the complex, heterogeneous nature of fibrotic lung tissue.

Future work could address these limitations by incorporating larger datasets from multiple animals and varying degrees of fibrosis to improve robustness and generalizability. Including additional biomarkers or multimodal imaging data could provide a more comprehensive representation of lung pathology. Extending the analysis to three dimensional reconstructions of lung tissue rather than individual slices would allow for more realistic spatial modeling. Finally, validating interpolation results against higher resolution sampling or alternative imaging modalities would be critical before translating this approach into clinical or device development applications.
